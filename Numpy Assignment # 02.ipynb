```python
#Task 16
def function16():
    #Write a NumPy program to join a sequence of arrays along depth.
    #x = np.array([[1], [2], [3]])
    #y = np.array([[2], [3], [4]])
    """
     Expected Output:
                [[[1 2]]
                 [[2 3]]
                 [[3 4]]]
     """ 

    x = np.array([[1], [2], [3]])
    y = np.array([[2], [3], [4]])
    ans = np.concatenate((x,y)).reshape((3,1,-1) , order="F")
    return ans

#or

#z=np.dstack((x,y))   (Ans with commas)
#z

#or

#z=np.concatenate([x,y],axis=1)   (Ans with commas)
#z

print(function16())
```
